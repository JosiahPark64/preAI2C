```python
import csv
import os


#open using fileIO
#create list of high temps
#calc alverage daily high temp for the year

high_temps = []
high_temps_flt=[]
with open('pittsburgh-weather-2024.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)

    for row in reader:
        row_list=row
        highest =row_list[1]
        #print(row_list)
        #print(row_list[1])
        
        #print(row['Day'], row['High'], row['Low'], row['Precipitation'], row['Snow'])
        #print(row)
        #hitemp = int(row.strip())
        high_temps.append(highest)

high_temps.pop(0)

for value in high_temps:
    high_temps_flt.append(float(value))
    
print(high_temps_flt)

avg_hi_temp = sum(high_temps_flt)/len(high_temps_flt)
print(avg_hi_temp)
```
